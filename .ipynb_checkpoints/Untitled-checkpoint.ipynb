<h2 style="color : red"><mark> Evlilik Statüsü doldurma aşaması </mark><h2>

```python
musteri_demografik['family_size'] = musteri_demografik.family_size.apply(lambda x: int(re.sub('\+','',x)))
musteri_demografik['no_of_children'] = musteri_demografik.no_of_children.apply(lambda x: int(re.sub('\+','',x)) if pd.notna(x) else x)
```

```python
#aile sayısı 1 olan müştrilerin hespi bekardır hem çocuk yok hem eş yok bu durumda bekardır 
musteri_demografik.loc[pd.isnull(musteri_demografik.marital_status) & (musteri_demografik.family_size == 1),'marital_status'] = 'Single'  # bu şekilde 183 kişi dolduruldu 

#aile sayısı - çocuk sayısı = 1 koşulunu sağlayan  müştrilerin hespi bekardır daha önce evlilik yapıp boşanmışlardır ve cocuk konusunda ya kendi çocugudur yada evlat edinmiştir haliyle bu aşamada eşi hesaba katmıyoruz 
musteri_demografik.loc[(musteri_demografik.family_size - musteri_demografik.no_of_children == 1) & pd.isnull(musteri_demografik.marital_status),'marital_status'] = 'Single'   # 33 deger dolduruldu 

# bir kişi evli ve aile sayısı iki ise çocuk sayısı 0 dır . mantıklı 
musteri_demografik.loc[pd.isnull(musteri_demografik.marital_status) & (pd.isnull(musteri_demografik.no_of_children))& (musteri_demografik.family_size ==2),'marital_status'] = 'Married'
```

<h2 style="color: red" ><mark>Cocuk doldurma aşaması</mark><h2>

```python
#evli insanların ailes sayısı iki ise ve ailede iki kişi avrsa çocuk yoktur 
musteri_demografik.loc[pd.isnull(musteri_demografik.no_of_children) & (musteri_demografik.marital_status == 'Married') & (musteri_demografik.family_size == 2),'no_of_children'] = 0

#Aile sayısı 1 olan müşteriler ise çocuk sahibi olmayacagından ötürü çcouk sayısı 0 olacaktır 
musteri_demografik.loc[pd.isnull(musteri_demografik.no_of_children) & (musteri_demografik.family_size == 1), 'no_of_children'] = 0
```

```python
#tabloda tam olarak yüzdesel görebilmek için eksik verileri tespit edip içlerine exsik veri yazdırdım. bu sayede
#histogram ve pasta grafiklerinde daha rahat bir şekilde analizini yapabileceğiz.

#imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='eksik_veri')
#musteri_demografik[["marital_status"]]=imputer.fit_transform(musteri_demografik[["marital_status"]])

#imputer2=SimpleImputer(missing_values=np.nan,strategy='constant', fill_value=99999)
#musteri_demografik[["no_of_children"]]=imputer2.fit_transform(musteri_demografik[["no_of_children"]])
```

```python
data = musteri_demografik.groupby("marital_status")["customer_id"].count()

f, ax = plt.subplots(3, 2, figsize=(15, 15))
 
f.suptitle('Musteri Demografik Özellikleri ve Histogramları')


ax[0,0].set_title('Yaş Guruplarındaki Kişi Sayıları', fontweight='bold')
ax[0,1].set_title('Evlilik Tiplerinin Yüzdelik Dagılımı', fontweight='bold')
ax[1,0].set_title('Rented-Müşteri Dağılımı (görüldüğü üzere bir outlier veri)', fontweight='bold')
ax[1,1].set_title('Aile Dagılımı', fontweight='bold')
ax[2,0].set_title('Çocuk Sayısı (0 ları kendimde doldurdum bu sayede sıfıların sayısını gözlemleyebiliriz)', fontweight='bold')
ax[2,1].set_title('Müşterilerimizin Gelir Dağılımları', fontweight='bold')

sns.countplot(ax=ax[0,0],x="age_range", data=musteri_demografik);
ax[0,1].pie(data.values,  labels=data.index, autopct='%1.1f%%')
sns.countplot(ax=ax[1,0],x="rented", data=musteri_demografik);
sns.countplot(ax=ax[1,1],x="family_size", data=musteri_demografik);
sns.countplot(ax=ax[2,0],x="no_of_children", data=musteri_demografik);
sns.countplot(ax=ax[2,1],x="income_bracket", data=musteri_demografik);
f.tight_layout()
plt.show()
```

Bu tablomuzda 7 adet sütunumuz bulunmaktadır. kolonların ne işe yaradığına gelecek olursak  
<br>

Bu aşamada gerekli (label / OneHot encoder) gibi dönüşümleri yaparak sayısal düzleme verimizi çekelim.

```python
encoder=LabelEncoder()
#Label Encoding Marital Status --- 0 is Single and 1 is Married
musteri_demografik[["marital_status"]] = encoder.fit_transform(musteri_demografik[["marital_status"]])


# Label Encoding age_range ... 18-25 is 0, 26-35 is 1, 36-45 is 2, 46-55 is 3, 56-70 is 4 and 70+ is 5
musteri_demografik[["age_range"]] = encoder.fit_transform(musteri_demografik[["age_range"]])
```

```python
#Xatam Değişkenimiz Zamana Çevrildi 
#If the DataFrame has a MultiIndex, this method can remove one or more levels.

musteri_islemleri['date'] = pd.to_datetime(musteri_islemleri['date'])
```

```python
#zamanı belirli periyotlarda tekrar edecek şekilde (günün saatleri haftanın günleri vb gibi) dönüşümler yapılarak yeni kolonlara yazdırıldı

musteri_islemleri['day_of_mounth'] = musteri_islemleri["date"].apply(lambda x: x.day)
musteri_islemleri['day_of_week'] = musteri_islemleri["date"].apply(lambda x: x.weekday())
musteri_islemleri['yer_of_mounth'] = musteri_islemleri["date"].apply(lambda x: x.month)
```

```python
musteri_islemleri
```

```python
# alınan ürünlerin tanesinin ne kadara geldiğini hesaplayalım
musteri_islemleri["indirimli_tane_fiyatı"] = (musteri_islemleri.selling_price)/(musteri_islemleri.quantity)

# ürünler indirime girmeseydi işlem ücretinin ne kadar olacagını hesaplayalım
musteri_islemleri["selling_price_indirimsiz"] = (musteri_islemleri.selling_price-musteri_islemleri.other_discount-musteri_islemleri.coupon_discount)

# ürünler indirime girmeseydi ürünlerin tanesinin ne kadar olacagını hesaplayalım
musteri_islemleri["indirimsiz_tane_fiyatı"] = (musteri_islemleri.selling_price-musteri_islemleri.other_discount-musteri_islemleri.coupon_discount)/(musteri_islemleri.quantity)
```

```python
# İşlemler gerçekleştiginde toplam ne kadarlik bir indirim oldugunu hesaplayalım 

musteri_islemleri["toplam_indirim"] = (musteri_islemleri.other_discount+musteri_islemleri.coupon_discount)
```
